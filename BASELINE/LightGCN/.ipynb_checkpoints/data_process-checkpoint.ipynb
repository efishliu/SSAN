```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df1 =  pd.read_csv("./Data/FilmTrust/ratings.txt",sep=' ',names=['userid','itemid','rating'])
```

```python
df1
```

```python
user_item_dict = {}
```

```python
for row in df1.itertuples():
    if user_item_dict.get(row.userid) is None:
        user_item_dict[row.userid] = []
    if float(row.rating) > 3.0:
        user_item_dict[row.userid].append(row.itemid)
    
    #print(row.userid,row.itemid,row.rating)
```

```python
user_item_dict
```

```python
i,j=0,0
user_map = {}
item_map = {}
data = []
```

```python
for k,v in user_item_dict.items():
    if len(v) > 0:
        if k not in user_map:
            user_map[k] = i
            i += 1
        for m in v:
            if m not in item_map:
                item_map[m] = j
                j += 1
            data.append([user_map[k],item_map[m]])
                
```

```python
data
```

```python
user_list=[]
item_list=[]
for k,v in user_map.items():
    user_list.append([k,v])
for k,v in item_map.items():
    item_list.append([k,v])
```

```python
df_user_list = pd.DataFrame(user_list,columns=["org_id","remap_id"])
```

```python
user_list
```

```python
df_user_list
```

```python
df_user_list.to_csv('./Data/FilmTrust/user_list.txt',index=0,sep=' ')
```

```python
df2 = pd.DataFrame(data,columns=['userid','itemid'])
```

```python
train_data,test_data = train_test_split(df2,test_size=0.2,random_state=42)
```

```python
train_data
```

```python
#train_data.to_csv('./Data/FilmTrust/train.txt',header=0,index=0,sep=' ')
```

```python
train = {}
test = {}
```

```python
for row in train_data.itertuples():
    if train.get(row.userid) is None:
        train[row.userid] = []
    train[row.userid].append(row.itemid)
```

```python
train
```

```python
with open('./Data/FilmTrust/train.txt','w') as f:
    for k,v in train.items():
        print(k,v)
        f.write(str(k))
        for m in v:
            f.write(' ')
            f.write(str(m))
        f.write('\n')
            
            
```
